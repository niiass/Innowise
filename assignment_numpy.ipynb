* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

- Task 1: Calculate the product of the non-zero elements on the diagonal of a rectangular matrix. For example, for X=np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]), the answer is 3.

```python
# your code
# WITHOUT NUMPY
x = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]
prod = 1
diag = [x[i][i] for i in range(min(len(x), len(x[0])))]
for num in diag:
    if num != 0: prod *= num
print(prod)

# WITH NUMPY
x = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
diag = np.diag(x)
print(np.prod(diag[diag!=0]))
```

- Task 2: Given two vectors x and y. Check if they define the same multiset. For example, for x=np.array([1, 2, 2, 4]) and y=np.array([4, 2, 1, 2]), the answer is True.

```python
# your code
# WITHOUT NUMPY
x = [1, 2, 2, 4]
y = [4, 2, 1, 2]
x_uniques = sorted(list(set(x)))
x_uniques_count = [x.count(i) for i in x_uniques]
y_uniques = sorted(list(set(y)))
y_uniques_count = [y.count(i) for i in y_uniques]
print((x_uniques == y_uniques) and (x_uniques_count == y_uniques_count))

# WITH NUMPY
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
x_uniques, x_uniques_count = np.unique(x, return_counts=True)
y_uniques, y_uniques_count = np.unique(y, return_counts=True)
print(len(x_uniques) == len(y_uniques) and np.all(x_uniques_count == y_uniques_count))
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль.

- Task 3: Find the maximum element in vector x among the elements that are preceded by a zero. For example, for x=np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]), the answer is 5.

```python
# your code
# WITHOUT NUMPY
x = [6, 2, 0, 3, 0, 0, 5, 7, 0]
zero_preceded = [x[i] for i in range(1, len(x)) if x[i-1] == 0]
print(max(zero_preceded))

# WITH NUMPY
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
zero_preceded = np.argwhere(x == 0) + 1
print(np.max(x[zero_preceded[zero_preceded < len(x)]]))
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

- Task 4: Implement Run-length encoding (RLE). For a given vector x, return a tuple of two vectors of equal length. The first vector contains the values, and the second contains the number of times each value is repeated consecutively. For example, for x=np.array([2, 2, 2, 3, 3, 3, 5]), the answer is (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# your code
# WITHOUT NUMPY
x = [2, 2, 2, 3, 3, 3, 5]
x_elem = []
x_elem_counts = []
count = 1
for i in range(len(x)):
    if (i + 1 < len(x)) and (x[i] == x[i+1]):
        count += 1
    else:
        x_elem.append(x[i])
        x_elem_counts.append(count)
        count = 1
print(x_elem, x_elem_counts)

# WITH NUMPY
x = np.array([2, 2, 2, 3, 3, 3, 5])
mask = x[1:] != x[:-1]
split_ind = np.concatenate(([0], np.where(mask)[0] + 1, [len(x)]))
print(x[split_ind[:-1]], np.diff(split_ind))
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

- Task 5: Given two sets of objects, X and y. Calculate the matrix of Euclidean distances between the objects. Compare your implementation with the scipy.spatial.distance.cdist function in terms of execution speed.

```python
# your code
import math
from scipy.spatial.distance import cdist
import time
# WITHOUT NUMPY
wo_np_start = time.time()
x = [[1, 2], [0, 3], [4, 6]]
y = [[2, 2], [3, 5]]
euclidean_distances = []

for point_x in x:
    for point_y in y:
        euclidean_distances.append(math.sqrt(pow((point_y[0]-point_x[0]), 2) + pow((point_y[1]-point_x[1]), 2)))
print(euclidean_distances)
wo_np_end = time.time()
print(f"\nExecution time for without numpy: {wo_np_end - wo_np_start}\n")

# WITH NUMPY
w_np_start = time.time()
x = np.array([[1, 2], [0, 3], [4, 6]])
y = np.array([[2, 2], [3, 5]])
x_grid = np.repeat(x, len(y), axis=0)
y_grid = np.tile(y, (len(x), 1))
diffs = y_grid - x_grid
print(np.linalg.norm(diffs, axis=1))
w_np_end = time.time()
print(f"\nExecution time for with numpy: {w_np_end - w_np_start}\n")

# WITH cdist
w_cdist_start = time.time()
print(cdist(x, y))
w_cdist_end = time.time()
print(f"\nExecution time with cdist: {w_cdist_end - w_cdist_start}\n")
```
