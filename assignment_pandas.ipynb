**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

8. Who has a higher proportion of high earners (>50K): married or unmarried men (the marital-status feature)? Consider as 'married' those whose marital-status starts with Married (Married-civ-spouse, Married-spouse-absent, or Married-AF-spouse); consider the rest as 'unmarried'.

```python
# your code
high_earners = data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]
high_earners_married = high_earners['marital-status'].str.startswith('Married')
married = data['marital-status'].str.startswith('Married')

high_earners_married_proportion = high_earners_married.sum()/married.sum()
high_earners_unmarried_proportion = (len(high_earners) - high_earners_married.sum())/(len(data)-married.sum())

print("Married men have higher proportion of high earners" if high_earners_married_proportion > high_earners_unmarried_proportion else "Unmarried men have higher proportion of high earners")
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

11. Group people into age categories young, adult, and retiree, where:

- young corresponds to 16–35 years

- adult corresponds to 35–70 years

- retiree corresponds to 70–100 years

Assign the name of the corresponding group to each person in a new column called 'AgeGroup'.

```python
# your code
ages = [16, 35, 70, 100]
categories = ['Young', 'Adult', 'Retiree']
data['AgeGroup'] = pd.cut(data['age'], bins=ages, labels=categories, include_lowest=True)
print(data)
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

12-13. Determine the number of people earning >50K in each of the age groups (the 'AgeGroup' column), and also output the name of the age group that most frequently earns more than 50K (>50K)

```python
# your code
age_categories = data.groupby(['AgeGroup', 'salary']).size()
categories_salary_count = age_categories.xs('>50K', level='salary')
print(f"Number of people earning >50K for each category:\n{categories_salary_count}\n")
print(f"Age group that most frequently earns >50k is {categories_salary_count.idxmax()}")
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

14. Group people by their type of employment (the occupation column) and determine the number of people in each group. Then, write a filtering function filter_func that returns only those groups where the average age (age column) is no more than 40 and in which every single worker works more than 5 hours per week (hours-per-week column).

```python
# your code
occupation_categories = data.groupby('occupation')
print(f"Number of people for each type of employment:\n{occupation_categories.size()}\n")

def filter_func(grouped_data):
    avg_age = grouped_data['age'].mean() <= 40
    hours_per_week = (grouped_data['hours-per-week'] > 5).all()
    return avg_age and hours_per_week

filtered_data = occupation_categories.filter(filter_func)
print(f"Filtered data with given limitations:\n{filtered_data}")
```
